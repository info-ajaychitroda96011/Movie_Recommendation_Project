### When we fetched genres[0] column ,It is in string format

### we need to convert this string format into list so we used ast module

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

### Apply convert Function into genres column

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head(2)
```

### Apply convert Function in keywords column

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head(2)
```

```python
movies['cast'][0]
```

### We need First three Data in cast column

```python
def convert_cast(obj):
    cast = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            cast.append(i['name'])
            counter+=1
        else:
            break
    return cast

```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
```

```python
movies.head(2)
```

```python
movies['crew'][0]
```

### We need only Director name so first we find job = 'Director' and than we fetched their name

```python
def convert_crew(obj):
    crew = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            crew.append(i['name'])
            break
        else:
            continue
    return crew
```

```python
movies['crew'] = movies['crew'].apply(convert_crew)
```

```python
movies.head(2)
```

### We Concatinate all the list and create one Tag

### But overview column contains string value so that first we changed its type from string to list and after than we concatinate all the list all together


```python
movies['overview'][0]
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head(2)
```

### NOTE: We remove all the spaces from the cast , crew and keywords columns

### Ex. We have two name:
- sam worthintug
- sam mendos

### if user will find sam worthintug movies but because of the spaces between every words there is a changed to make mistake 

### User will find sam worthintug movie and he find sam mendos movie instead

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","")for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movies.head(2)
```

### Create Tag column and Bind the all the columns(genres,keywords,cast,crew)

```python
movies['tag'] = movies['overview'] + movies['genres']  + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head(2)
```

```python
new_df = movies[['movie_id','title','tag']]

new_df.head()
```

### Convert tag column from list into string

```python
new_df['tag'] = new_df['tag'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df['tag'] = new_df['tag'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
import numpy as np
```

```python
vectors = cv.fit_transform(new_df['tag']).toarray()
```

```python
cv.get_feature_names_out()
```

```python
import nltk
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer

wl = WordNetLemmatizer()
```

```python
def lemmatize(text):
    y = []

    for i in text.split():
        y.append(wl.lemmatize(i))

    return " ".join(y)
```

```python
new_df['tag'] = new_df['tag'].apply(lemmatize)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cvv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cvv.fit_transform(new_df['tag']).toarray()
```

```python
np.set_printoptions(threshold=np.inf)
```

```python
cvv.get_feature_names_out()
```

### We calculate distance of one vector with another vector

### We have 5000 Dimension vectors so we used cosine similarity


```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
print(similarity[0])
```

### Now we fetched top 5 movie with similar score

```python
new_df[new_df['title'] == 'Avatar'].index[0]
```

### When we sorted our movie at that time we will loose movie index so that we perform below step

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movies):
    movies_index = new_df[new_df['title'] == movies].index[0]
    distances = similarity[movies_index]
    movie_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

    for i in movie_list:
        print(new_df.iloc[i[0]].title)
        # print(i[0])
```

```python
recommend('Batman Begins')
```
